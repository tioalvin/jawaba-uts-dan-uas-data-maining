```python
#soal pertama Lakukan analisa mengenai field mana saja yang memiliki data kosong(Nan) lalu Isi data yang kosong tersebut menggunakan nilai mean

import pandas as pd

# Membaca file CSV
data_startup = pd.read_csv('50_Startups.csv')

# Menemukan nilai kosong (NaN) untuk setiap kolom numerik
missing_values = data_startup.select_dtypes(include='number').isna().sum()
print("Jumlah data kosong (NaN) untuk setiap kolom:\n", missing_values)

# Mengisi nilai kosong dengan nilai mean untuk setiap kolom numerik
data_startup.fillna(data_startup.select_dtypes(include='number').mean(), inplace=True)

# Menampilkan data setelah diisi nilai mean
print("\nData setelah mengisi data kosong dengan nilai mean:\n", data_startup.sample(7))
```

```python
#soal 2 lakukan oneHotEncoder ke field state

import pandas as pd

# Membaca file CSV
data_startup = pd.read_csv('50_Startups.csv')

# Melakukan one-hot encoding pada kolom 'State'
data_startup_encoded = pd.get_dummies(data_startup, columns=['State'])

# Menampilkan data setelah one-hot encoding
print("Data setelah one-hot encoding:\n", data_startup_encoded.sample(7))
```

```python
#soal 3 buat field baru yaitu Tax. dimana Tax = (profit + Marketing Spend + administration) * 5%

import pandas as pd

# Membaca file CSV
data_startup = pd.read_csv('50_Startups.csv')

# Membuat kolom baru 'Tax'
data_startup['Tax'] = (data_startup['Profit'] + data_startup['Marketing Spend'] + data_startup['Administration']) * 0.05

# Menampilkan data setelah menambahkan kolom 'Tax'
print("Data setelah menambahkan kolom 'Tax':\n", data_startup.sample(7))
```

```python
# Soal 4 Jalankan standardScaler field-field tersebut

import pandas as pd
from sklearn.preprocessing import StandardScaler

# Membaca file CSV
data_startup = pd.read_csv('50_Startups.csv')

# Membuat kolom baru 'Tax'
data_startup['Tax'] = (data_startup['Profit'] + data_startup['Marketing Spend'] + data_startup['Administration']) * 0.05

# Mengambil fitur-fitur yang akan di-scaling
features = data_startup.drop(['State', 'Profit'], axis=1)

# Melakukan scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Mengganti nilai fitur-fitur dengan nilai yang telah di-scaling
data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'Tax']] = scaled_features

# Menampilkan data setelah scaling
print("Data setelah scaling:\n", data_startup.sample(7))

```
